```python
final = df.drop(['X1 transaction date','No'],axis = 1)
final
```

```python
#define input and output
X = final.drop(columns='areaprice')
y = final.iloc[:, 5]
y
```

```python
X
```

```python
#normalization #X-min/max-min
data = ((X-X.min())/(X.max()-X.min()))
data
```

```python
#bias term
d = pd.concat([pd.Series(1,index=data.index,name='00'),data],axis=1)
```

```python
d
```

```python
#Initialize theta value
theta = np.array([0]*len(d.columns)) #Output: array([0,0,0,0,0,0])
```

```python
#length of rows
m = len(d)
m
```

```python
#hypothesis function
def hypothesis(theta, d):
    return theta*d
```

```python
#compute cost
#RMSE
def computeCost(d, y, theta):
    y1 = hypothesis(theta, d) #y1 = predicted value
    y1 = np.sum(y1, axis = 1) 
    return sum(np.sqrt((y1-y)**2))/(2*m) # y = actual value
```

```python
def gradientDescent(d, y, theta, alpha, i):
    J = [] #cost function in each iterations
    k = 0
    while k < i:
        y1 = hypothesis(theta, d)
        y1 = np.sum(y1, axis=1)
        for c in range(0, len(d.columns)):
            theta[c] = theta[c] - alpha*(sum((y1-y)*d.iloc[:,c])/len(d))
        j = computeCost(d, y, theta)
        J.append(j)
        k +=1
    return J, j, theta
```

```python
#call gradient descent
J, j, theta = gradientDescent(d, y, theta, 0.05, 10000)
```

```python
#call hypothesis
y_hat = hypothesis(theta, d)
y_hat = np.sum(y_hat, axis = 1)
```
